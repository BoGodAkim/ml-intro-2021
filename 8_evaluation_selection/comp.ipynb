```python
task = task[['taskId','courseTaskId','maxScore','scoreWeight','courseId']]
task
```

```python
#task_res = task_res.drop_duplicates(subset=['studentId','courseTaskId'])
task_res = task_res[['score','studentId','courseTaskId']]
for i in task[task['maxScore'].isnull()]['courseTaskId'].values:
    task.loc[task['courseTaskId']==i,'maxScore']=task_res[task_res['courseTaskId']==i]['score'].max()
task_res
```

```python
scores = {}
max_scor_loc = {}
for i in task_res['studentId'].value_counts().index:
    for o in task_res[task_res['studentId']==i]['courseTaskId'].value_counts().index:
        if i in scores:
            scores[i]=scores[i]+task_res[(task_res['studentId']==i) & (task_res['courseTaskId']==o)]['score'].mean()*task[task['courseTaskId']==o]['scoreWeight'].sum()
            max_scor_loc[i]=max_scor_loc[i]+task[task['courseTaskId']==o]['maxScore'].sum()*task[task['courseTaskId']==o]['scoreWeight'].sum()
        else:
            scores[i]=task_res[(task_res['studentId']==i) & (task_res['courseTaskId']==o)]['score'].mean()*task[task['courseTaskId']==o]['scoreWeight'].sum()
            max_scor_loc[i]=task[task['courseTaskId']==o]['maxScore'].sum()*task[task['courseTaskId']==o]['scoreWeight'].sum()
        #scores[2].append(task_res[(task_res['studentId']==i) & (task_res['courseTaskId']==o)]['score'].sum())
```

```python
max_scor={}
for i in task['courseTaskId'].value_counts().index:
    c=task[task['courseTaskId']==i]['courseId'].sum()
    if c in max_scor:
        max_scor[c] = max_scor[c]+task[task['courseTaskId']==i]['maxScore'].sum()*task[task['courseTaskId']==i]['scoreWeight'].sum()
    else:
        max_scor[c] = task[task['courseTaskId']==i]['maxScore'].sum()*task[task['courseTaskId']==i]['scoreWeight'].sum()
```

```python
train['score']=0
for i in train['studentId']:
    if i in scores:
        train.loc[train['studentId']==i,'score']=scores[i]/max_scor[train[train['studentId']==i]['courseId'].sum()]
    else:
        train.loc[train['studentId']==i,'score']=0
test['score']=0
for i in test['studentId']:
    if i in scores:
        test.loc[test['studentId']==i,'score']=scores[i]/max_scor[test[test['studentId']==i]['courseId'].sum()]
    else:
        test.loc[test['studentId']==i,'score']=0
```

```python
train['score_loc']=0
for i in train['studentId']:
    if i in scores:
        train.loc[train['studentId']==i,'score_loc']=scores[i]/max_scor_loc[i]
    else:
        train.loc[train['studentId']==i,'score_loc']=0
test['score_loc']=0
for i in test['studentId']:
    if i in scores:
        test.loc[test['studentId']==i,'score_loc']=scores[i]/max_scor_loc[i]
    else:
        test.loc[test['studentId']==i,'score_loc']=0
```

```python
train.loc[train['score_loc']>1,'score_loc']=1
test.loc[test['score_loc']>1,'score_loc']=1
```

```python
def save(y,ind,name='out.csv'):
    out=[]
    for i in range(len(y)):
        if y[i]==1:
            out.append(True)
        else:
            out.append(False)
    result = pd.DataFrame({'studentId': ind, 'isTopPerformer': out})
    result.to_csv(name, index=False)
    return 'Done'
```
